```python
import pandas as pd
df=pd.read_csv("prices.csv")
df
```

```python
d=pd.get_dummies(df.town)
```

```python
merged=pd.concat([df,d],axis="columns")
merged
```

```python
merged["monroe township"]=merged["monroe township"].astype(int)
merged["robbinsville "]=merged["robbinsville "].astype(int)
merged["west windsor"]=merged["west windsor"].astype(int)
merged
```

```python
final=merged.drop(["town","west windsor"],axis="columns")
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop("price",axis="columns")
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
